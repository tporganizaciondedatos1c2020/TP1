```python

top_loc = list(dataset.groupby('location').agg({'target':'count'}).nlargest(20, 'target').reset_index().location)
top_loc
```

```python
grouped = dataset.loc[(dataset['location'].isin(top_loc)),:]
grouped = grouped.groupby(['location', 'longitud']).agg({'target':'count'}).reset_index()
grouped.columns = ['Location', 'Longitud', 'Cantidad']
grouped.dropna(inplace=True)
grouped
```

```python
heatmap_data = pd.pivot_table(grouped, values='Cantidad', 
                     index=['Location'], 
                     columns='Longitud')
heatmap_data = heatmap_data.fillna(0)
heatmap_data
```
